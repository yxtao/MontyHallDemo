```python
def switchChoice(choices,change):
    if(change==True) :
        return choices[1];
        
    else :
        return choices[0];
```

```python
def playGame(doorNum, change,choices):
    #get user' choice Item
    choiceItem= choices[doorNum-1];
    # create an new array after removing a "goat" element
    choicesAfterRemove =["car","goat"];
    if (choiceItem=="car"):
        #removed a "goat",left "car" and "goat"
        choicesAfterRemove[0]= "car";
        choicesAfterRemove[1]="goat";
    else : #when the user chooses a "goat"
        #removed a "goat", left "car' and "goat"
        choicesAfterRemove[0]= "goat";
        choicesAfterRemove[1]="car";
        # the choicesAfterRemove array contains two elements,
        # the first element is the original chose item
        #call makeChoice helper method
    finalItem = switchChoice(choicesAfterRemove,change);

    return finalItem;
```

```python
def randomChoices():
    import random
    value = random.randint(0,2)
    if (value==0):
        return ["car","goat","goat"]
    elif (value==1):
        return ["goat","car","goat"]
    else:
        return ["goat","goat","car"]
  
```

```python
# doorNum 0,1,2, 
# times: how many times playing the game
# change: True if switch
def games(times, doorNum, change):
    countCar=0
    countGoat=0
    for x in range(times):
        choices=randomChoices();
        final=playGame(doorNum,change,choices)
        if(final=="car") :
            countCar= countCar+1
        else :
            countGoat= countGoat+1
    return countCar,countGoat
            
```

```python
# switch 
result=games(100,1,True) # (timesOfTheGame, doorNum0,1 or2, switchOrNot)
```

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'cars', 'goats'
sizes = result
explode = (0, 0.1, )  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
# do not switch 

result=games(100,1,False)
```
